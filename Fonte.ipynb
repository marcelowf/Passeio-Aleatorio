```python
import matplotlib.pyplot as plt
import numpy as np

def hex_random_walk(n):
    directions = np.array([
        [1, 0],
        [0.5, np.sqrt(3)/2],
        [-0.5, np.sqrt(3)/2],
        [-1, 0],
        [-0.5, -np.sqrt(3)/2],
        [0.5, -np.sqrt(3)/2]
    ])

    position = np.array([0.0, 0.0])
    walk = [position.copy()]

    for _ in range(n):
        step = directions[np.random.randint(0, 6)]
        position += step
        walk.append(position.copy())
    return np.array(walk)
n = 24

walk = hex_random_walk(n)

plt.figure(figsize=(10, 10))
plt.plot(walk[:, 0], walk[:, 1], marker='o')
plt.title(f'Passeio Aleatrio Hexagonal com {n} Passos')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
